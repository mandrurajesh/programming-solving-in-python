## Adding of data

### Problem 2:

#### Problem Statement: Adding data to the  csv file

#### CONSTRAINTS: CSV file 

#### Test cases:
* Test case 1
* Test case 2
* Test case 3

```python
import pandas as pd
def add_row_csv(filename,rowdata):
    df=pd.read_csv(filename)
    df.loc[len(df)]=rowdata
    df.to_csv(filename,index=False)
    return df
rowdata=['010101','statement',123,456,567,356,467,868,468,688,789]
add_row_csv('DataFiles/income.csv',rowdata)
```

## Deleting a row from the csv file

### Problem 4:

#### Problem Statement: Deleting the data from csv file

#### CONSTRAINTS: CSV file 

#### Test cases:
* Test case 1
* Test case 2
* Test case 3

```python
def deleteRowcsv(filename,rowindex):
    df=readcsv(filename)
    df=df.drop(rowindex)
    df.to_csv(filename,index=False)
    return df
```

```python
filename='DataFiles/income.csv'
rowindex=7
deleteRowcsv(filename,rowindex)
```
